```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(tips.sex,tips.day)
plt.title("Tip vs Size")
plt.xlabel("Tip")
plt.ylabel("Size")
```

```python
pd.crosstab(tips.sex,tips.day).plot(kind="bar")
```

```python
tips.day.value_counts().plot(kind='pie')
```

```python
plt.hist(tips.tip)
```

```python
plt.boxplot(tips['size'])
```

```python
plt.violinplot(tips['size'])
```

```python
sns.stripplot(x='day',y='size',data=tips)
plt.show()
```

```python
sns.swarmplot(x='tip',y='time',data=tips)
plt.show()
```

```python
sns.jointplot(x='tip',y='time',data=tips)
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sor=tips.corr()
sns.heatmap(sor)
```

```python
sns.violinplot(x='day',y='tip',data=tips)
plt.show()
```
