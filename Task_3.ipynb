```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols("AT~Waist", data=data).fit()
```

```python
model.params
```

```python
model = smf.ols("Waist~AT", data=data).fit()
```

```python
model.params
```

```python
model.rsquared
```
